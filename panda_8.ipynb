### Indexing---
#### index-based selection: selecting data based on its numerical position in the data

```python
df.iloc[0]
```

```python
df.iloc[:,3]
```

```python
df.iloc[0:3,1]
```

```python
# pass a list
df.iloc[[0,2,4],2]
```

```python
df.iloc[-5:]
```

### label based selection loc()

```python
df.loc[0:5,"sex"]
```

```python
df.loc[:,["sex","total_bill","day"]]
```

## conditional selection

```python
df.sex == "Male"
```

```python
df[df.sex == "Male"]
```

```python
df.loc[df.sex == "Male"]
```

```python
df.loc[df.sex == "Male",["sex","smoker","total_bill"]]
```

```python
df.loc[df.day.isin(["Sun","Sat"])]
```
