### nth highest salary

```python
import pandas as pd
```

```python
employee_table1 = pd.DataFrame({
    "id":[1,2,3],
    "salary":[100,200,300]
})
```

```python
employee_table1
```

```python
sorted_emply = employee_table1.salary.drop_duplicates().sort_values(ascending=False)
```

```python
sorted_emply
```

```python
n = 2
```

```python
len(sorted_emply)
```

```python
if n > len(sorted_emply):
    nth_highest = None
else:
    nth_highest = sorted_emply.iloc[n-1]
```

```python
nth_highest
```

```python
res_col_name=f'getNthHighestSalary({n})'
res_col_name
```

```python
res_df = pd.DataFrame({
    res_col_name:[nth_highest]
})
```
