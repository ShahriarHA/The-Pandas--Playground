```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

## Grouping and sorting

```python
df.day.value_counts()
```

```python
# alternative of value_counts()
df.groupby("day").day.count()
```

```python
df.groupby("day").total_bill.min()
```

```python
df.groupby("day").total_bill.max()
```

```python
df.total_bill.idxmax() # return max value index
```

```python
df.total_bill.max()
```

```python
df.iloc[170]
```

```python
df.groupby("day").total_bill.agg([len,min,max,"mean","median"])
```

```python
df.groupby(["day","sex"]).sex.agg([len])
```

```python
df.index
```

```python
day_by = df.groupby(["day","sex"]).sex.agg([len]).reset_index()
```
