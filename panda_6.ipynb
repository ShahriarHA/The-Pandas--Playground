```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data_wide = {
    'Salesperson': ['Alice', 'Bob', 'Charlie'],
    'Apples': [330, 0, 0],
    'Bananas': [0, 370, 0],
    'Oranges': [0, 0, 175]
}
```

```python
df = pd.DataFrame(data_wide)
```

```python
df
```

```python
df.info()
```

```python
melted_df = pd.melt(df,id_vars="Salesperson",value_vars=["Apples","Bananas","Oranges"],var_name="Product",value_name="Sales")
```

```python
melted_df
```

## melt() turns columns into rows
## When to use melt():
#### 1. When preparing data for visualization (most plotting libraries prefer long format)
#### 2. Before using groupby operations on what were previously column names
