## 1873. Calculate Special Bonus Leetcode problem

```python
Employees = pd.DataFrame({
    'employee_id': [2, 3, 7, 8, 9],
    'name': ['Meir', 'Michael', 'Addilyn', 'Juan', 'Kannon'],
    'salary': [3000, 3800, 7400, 6100, 7700]
})
```

```python
Employees
```

```python
Employees[Employees.name.str.startswith("M")]
```

```python
Employees[~Employees.employee_id % 2 == 0]
```

```python
Employees.loc[(~Employees.employee_id%2==0)&(~Employees.name.str.startswith("M"))]
```

```python
eligible_for_bonus = Employees[(~Employees.employee_id%2==0)&(~Employees.name.str.startswith("M"))]
```

```python
eligible_for_bonus
```

```python
import numpy as np
```

```python
condition = ((~Employees.employee_id%2==0)&(~Employees.name.str.startswith("M")))
Employees["bonus"] = np.where(condition,Employees.salary,0)
```

```python
Employees
```

```python
Employees[["employee_id","bonus"]].sort_values(by="employee_id")
```
