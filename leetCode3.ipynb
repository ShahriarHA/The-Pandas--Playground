### Department highest salary

```python
employee_data = {
    'id': [1, 2, 3, 4, 5],
    'name': ['Joe', 'Jim', 'Henry', 'Sam', 'Max'],
    'salary': [70000, 90000, 80000, 60000, 90000],
    'departmentId': [1, 1, 2, 2, 1]
}

employee = pd.DataFrame(employee_data)

department_data = {
    'id': [1, 2],
    'name': ['IT', 'Sales']
}
department = pd.DataFrame(department_data)
```

```python
employee
```

```python
employee.groupby("departmentId")["salary"].sum()
```

```python
department
```

```python
df = employee.merge(department,left_on="departmentId",right_on="id",suffixes=("_em","_de"))
df
```

```python
max_salary = df.groupby("name_de")["salary"].agg(max)
max_salary.reset_index()
```

```python
max_salary_re = max_salary.reset_index()
max_salary_re
```

```python
result = df.merge(max_salary_re,on=["name_de","salary"])
result
```

```python
result[["name_de","name_em","salary"]].set_axis(["Department","Employee","Salary"],axis=1)
```

### Rank Scores

```python
data = {
    'id': [1, 2, 3, 4, 5, 6],
    'score': [3.50, 3.65, 4.00, 3.85, 4.00, 3.65]
}
scores_df = pd.DataFrame(data)
scores_df
```

```python
sorted_scores_df = scores_df.sort_values(by="score",ascending=False)
sorted_scores_df
```

```python
sorted_scores_df["rank"] = sorted_scores_df.score.rank(method='dense',ascending=False)
sorted_scores_df
```

```python
sorted_scores_df["rank"] = sorted_scores_df["rank"].astype(int)
```

```python
sorted_scores_df
```

```python
sorted_scores_df[["score","rank"]]
```
