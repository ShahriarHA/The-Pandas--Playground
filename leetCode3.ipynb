### Department highest salary

```python
employee_data = {
    'id': [1, 2, 3, 4, 5],
    'name': ['Joe', 'Jim', 'Henry', 'Sam', 'Max'],
    'salary': [70000, 90000, 80000, 60000, 90000],
    'departmentId': [1, 1, 2, 2, 1]
}

employee = pd.DataFrame(employee_data)

department_data = {
    'id': [1, 2],
    'name': ['IT', 'Sales']
}
department = pd.DataFrame(department_data)
```

```python
employee
```

```python
employee.groupby("departmentId")["salary"].sum()
```

```python
department
```

```python
df = employee.merge(department,left_on="departmentId",right_on="id",suffixes=("_em","_de"))
df
```

```python
max_salary = df.groupby("name_de")["salary"].agg(max)
max_salary.reset_index()
```

```python
max_salary_re = max_salary.reset_index()
max_salary_re
```

```python
result = df.merge(max_salary_re,on=["name_de","salary"])
result
```

```python
result[["name_de","name_em","salary"]].set_axis(["Department","Employee","Salary"],axis=1)
```

### Rank Scores

```python
data = {
    'id': [1, 2, 3, 4, 5, 6],
    'score': [3.50, 3.65, 4.00, 3.85, 4.00, 3.65]
}
scores_df = pd.DataFrame(data)
scores_df
```

```python
sorted_scores_df = scores_df.sort_values(by="score",ascending=False)
sorted_scores_df
```

```python
sorted_scores_df["rank"] = sorted_scores_df.score.rank(method='dense',ascending=False)
sorted_scores_df
```

```python
sorted_scores_df["rank"] = sorted_scores_df["rank"].astype(int)
```

```python
sorted_scores_df
```

```python
sorted_scores_df[["score","rank"]]
```

### Count Salary Ctegories

```python
accounts = pd.DataFrame({
    'account_id': [3, 2, 8, 6],
    'income': [108939, 12747, 87709, 91796]
})
accounts
```

```python
# def count_category(x):
#     if x < 20000:
#         return 
#     # elif :
#     #     return "Average Salary"
#     # else:
#     #     return "High Salary"
```

```python
# accounts["category"] = accounts.income.apply(select_category)
accounts["Low Salary"] = accounts.income.apply(lambda x: 1 if x < 20000 else 0)
accounts["Average Salary"] = accounts.income.apply(lambda x: 1 if (x >= 20000) and (x <= 50000) else 0)
accounts["High Salary"] = accounts.income.apply(lambda x: 1 if x > 50000 else 0)
```

```python
accounts
```

```python
melted_account = pd.melt(accounts,id_vars="income",value_vars=["Low Salary","Average Salary","High Salary"],var_name="category",value_name="accounts_count")
```

```python
melted_account
```

```python
res = melted_account.groupby("category")["accounts_count"].agg(sum).reset_index()
```
