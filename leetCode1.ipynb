### find users with valid emails

```python
df2 = pd.DataFrame({
    'user_id': [1, 2, 3, 4, 5, 6, 7],
      'name': ['Winston', 'Jonathan', 'Annabelle', 'Sally', 'Marwan', 'David', 'Shapiro'],
      'mail': [
          'winston@leetcode.com',
          'jonathanisgreat',
          'bella-@leetcode.com',
          'sally.come@leetcode.com',
          'quarz#2020@leetcode.com',
          'david69@gmail.com',
          '.shapo@leetcode.com'
      ]
})
```

```python
df2
```

```python
regex_pattern = r'^[a-zA-Z][a-zA-Z0-9._-]*@leetcode[.]com$'
```

```python
df2[df2.mail.str.match(regex_pattern)]
```

### patients with condition

```python
df3 = pd.DataFrame({
    'patient_id': [1, 2, 3, 4, 5],
      'patient_name': ['Daniel', 'Alice', 'Bob', 'George', 'Alain'],
      'conditions': [
          'YFEV COUGH',
          '',  # Represents the empty string condition for Alice
          'DIAB100 MYOP',
          'ACNE DIAB100',
          'DIAB201'
      ]
})
```

```python
df3
```

```python
pattern = r'^DIAB1| DIAB1'
```

```python
df3[df3.conditions.str.contains(pattern)]
```

### Delete Duplicate Emails

```python
df4 = pd.DataFrame({
    "id":[1,2,3],
    "email": ["john@example.com","bob@example.com","john@example.com"]
})
```

```python
df4
```

```python
df4.email.duplicated()
```

```python
df4.drop_duplicates(subset=['email'],keep='first',inplace=True)
```

### Rearrange products table

```python
import numpy
```

```python
product_table = pd.DataFrame({
    "product_id":[0,1],
    "store1":[95,70],
    "store2":[100,numpy.nan],
    "store3":[105,80],

})
```

```python
product_table
```

```python
pt = pd.melt(product_table,id_vars="product_id",value_vars=["store1","store2","store3"],var_name="store",
        value_name="price").sort_values(by="product_id")
```

```python
pt
```

```python
pt.columns
```

```python
pt.dropna(subset="price",inplace=True)
```
