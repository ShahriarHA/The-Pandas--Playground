```python
import pandas as pd
```

```python
Customers = pd.DataFrame({
    "id": [1,2,3,4],
    "name": ["Joe","Henry","Sam","Max"]
})
Orders = pd.DataFrame({
    "id": [1,2],
    "customer_id": [3,1]
})
```

```python
Customers
```

```python
Orders
```

```python
merged = pd.merge(Customers,Orders,on="id",how="inner")
```

```python
merged
```

```python
pd.merge(Orders,Customers,on="id",how="inner")
```

### Leetcode probelm with isin()

```python
Customers[Customers.id.isin(Orders.customer_id)]
```

```python
isin = Customers[~Customers.id.isin(Orders.id)]
```

```python
isin
```

```python
isin1 = isin.rename(columns={
    "name": "customers"
})
```

```python
isin1
```

```python
pd.DataFrame(isin1.customers)
```

```python
df1 = pd.DataFrame({'customerID': [1, 2, 3, 4], 'name': ['A', 'B', 'C', 'D']})
df2 = pd.DataFrame({'ID': [1, 2], 'customerID': [2,4]})
```

```python
pd.merge(df1,df2,on="customerID",how="left")
```

```python
pd.merge(df1,df2,on="customerID",how="right")
```
