### Replacing Null values with mean, median, mode

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_missing_values = df.loc[df.isnull().any(axis=1)]
```

```python
# bill_length_mm with median values
bill_length_mm_median = df.bill_length_mm.median()
bill_length_mm_median
```

```python
df_missing_values.fillna({"bill_length_mm":bill_length_mm_median},inplace=True)
```

```python
df_missing_values
```

```python
# bill_depth_mm with median values
bill_depth_mm_meian = df.bill_depth_mm.median()
df_missing_values.fillna({"bill_depth_mm":bill_depth_mm_meian},inplace=True)
```

```python
df_missing_values
```

```python
# flipper_length_mm with mean
flipper_length_mm_mean = df.flipper_length_mm.mean()
```

```python
flipper_length_mm_mean
```

```python
df_missing_values.fillna({"flipper_length_mm":flipper_length_mm_mean},inplace=True)
```

```python
df_missing_values
```

```python
# body_mass_g with median
body_mass_g_median = df.body_mass_g.median()
```

```python
body_mass_g_median
```

```python
df_missing_values.fillna({"body_mass_g":body_mass_g_median},inplace= True)
```

```python
df_missing_values
```

```python
# dealing "sex" column with mode value
sex_mode = df.sex.mode()[0]
```

```python
sex_mode
```

```python
df_missing_values.fillna({"sex":sex_mode},inplace=True)
```

```python
df_missing_values
```
